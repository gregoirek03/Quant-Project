```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import time
from performance_tools import calculate_moving_average_vectorized
```

```python
# Simulation de 100 000 prix (ex: Apple)
rng = np.random.default_rng(seed=42)
prix = rng.uniform(170, 180, 100000)
fenetre = 20

# 1. Approche lente (La boucle for)
start_for = time.time()
moyennes_boucle = []
for i in range(len(prix)):
    if i < fenetre:
        moyennes_boucle.append(np.nan)
    else:
        moyennes_boucle.append(np.mean(prix[i-fenetre:i]))
end_for = time.time()
temps_for = end_for - start_for

# 2. Approche rapide (Ta fonction vectorisée)
start_vec = time.time()
moyennes_vectorisees = calculate_moving_average_vectorized(prix, fenetre)
end_vec = time.time()
temps_vec = end_vec - start_vec

print("="*50)
print(f"Temps avec boucle for : {temps_for:.4f} secondes")
print(f"Temps vectorisé :        {temps_vec:.4f} secondes")
print(f"Gain de rapidité :       x{int(temps_for / temps_vec)}")
print("="*50)
```
