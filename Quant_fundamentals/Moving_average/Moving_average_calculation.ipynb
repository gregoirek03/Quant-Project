```python
# Pour optimiser du code Python en finance --> vectorisation 
# adieu boucles for et laisser NumPy et Pandas faire les calculs en C 

# ex : calculer une moyenne glissante

# ❌ L'approche "Non-Quant" (La boucle for)
# On parcourt les prix un par un
# C'est très lent dès qu'il y a des milliers de lignes.

import time

# Simulation de 100 000 prix d'Apple
import numpy as np
import pandas as pd

prix = np.random.uniform(170, 180, 100000)
fenetre = 20
moyennes_boucle = []

start = time.time()
# On boucle sur chaque élément... l'enfer des performances
for i in range(len(prix)):
    if i < fenetre:
        moyennes_boucle.append(np.nan)
    else:
        moyennes_boucle.append(np.mean(prix[i-fenetre:i]))
end = time.time()

print(f"Temps avec boucle for : {end - start:.4f} secondes")
```

```python
# L'approche "Quant" (Vectorisée avec Pandas)
# on utilise les fonctions natives codées en C
# --> traite les données comme un bloc (un vecteur), pas une par une

df = pd.DataFrame({'Prix': prix})

start = time.time()
# Une seule ligne, ultra optimisée
df['Moyenne_Rolling'] = df['Prix'].rolling(window=fenetre).mean()
end = time.time()

print(f"Temps vectorisé : {end - start:.4f} secondes")


# .rolling(window=fenetre) : 
# "Rolling" signifie glissant. 
# Pandas va créer des sous-groupes de données. 
# Si ta fenetre est réglée sur 20, Pandas se place sur la 20ème ligne et regarde les 20 dernières lignes. 
# Puis il passe à la 21ème ligne, et regarde les 20 dernières, et ainsi de suite.

# .mean() : 
# On dit à Pandas : "Sur chaque sous-groupe de 20 prix que tu viens de cibler, calcule la moyenne."

# df['Moyenne_Rolling'] = ... : 
# On crée une toute nouvelle colonne dans notre tableau pour y stocker directement tous les résultats d'un coup.
```
