```python
import numpy as np

# Même programme que pour logistique

# 1. CONFIGURATION ET GÉNÉRATION DES DONNÉES 
np.random.seed(42)
n_samples = 100
n_features = 2

# Génération des features X et ajout de l'intercept X_b
X = np.random.randn(n_samples, n_features)
X_b = np.c_[np.ones((n_samples, 1)), X]

# Coefficients théoriques secrets et fonction sigmoïde
true_beta = np.array([[0.5], [2.5], [-1.5]])

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Génération de la cible Y (0 ou 1)
probabilities = sigmoid(np.dot(X_b, true_beta))
Y = (probabilities > 0.5).astype(int)

# 2. DESCENTE DE GRADIENT
beta = np.zeros((n_features + 1, 1))
learning_rate = 0.1
n_iterations = 1000

for i in range(n_iterations):
    Z = np.dot(X_b, beta)
    Y_pred = sigmoid(Z)
    
    # Calcul du gradient et mise à jour
    gradient = (1 / n_samples) * np.dot(X_b.T, (Y_pred - Y))
    beta = beta - learning_rate * gradient
    
# Classification 

# 3. ÉVALUATION ET MATRICE DE CONFUSION 
Z_final = np.dot(X_b, beta)
Y_prob = sigmoid(Z_final)
Y_pred_class = (Y_prob >= 0.5).astype(int)

y_true = Y.ravel()
y_pred = Y_pred_class.ravel()

# Calcul des 4 cases
TP = np.sum((y_true == 1) & (y_pred == 1))
TN = np.sum((y_true == 0) & (y_pred == 0))
FP = np.sum((y_true == 0) & (y_pred == 1))
FN = np.sum((y_true == 1) & (y_pred == 0))

accuracy = (TP + TN) / len(y_true)

# Affichage du bilan complet
print("=== RÉSULTATS DE LA CLASSIFICATION ===")
print(f"Accuracy : {accuracy * 100:.2f}%\n")
print("Matrice de Confusion :")
print(f"               Prédit 0   Prédit 1")
print(f"Réellement 0 |    {TN}     {FP}")
print(f"Réellement 1 |    {FN}     {TP}")
```
