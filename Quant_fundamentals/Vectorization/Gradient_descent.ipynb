```python
# 🏋️ Exercice 2 : La mise à jour du Gradient (Descente de Gradient)
# On cherche à faire une régression linéaire standard y = X*beta
# La matrice X est trop gigantesque pour être inversée directement
# --> descente de gradient
# --> Écrire la fonction qui calcule le gradient de l'erreur quadratique par rapport aux coefficients beta.

# 🧠 L'explication mathématique
# Fonction de coût
# J(beta) = 1/N * ||X*beta - y\||^2.
# Si on dérive par rapport à beta, on obtient la formule du gradient 
# Gradient = 2/N * X^T*(X*beta - y)
# --> (X*beta - y) représente le vecteur de toutes les erreurs de prédiction
# --> On multiplie par X^T pour projeter cette erreur sur chaque variable.
```

```python
def calculer_gradient(X: np.ndarray, y: np.ndarray, beta: np.ndarray) -> np.ndarray:
    """
    Calcule le gradient pour une régression linéaire de manière 100% vectorisée.
    X: (N, P) - les features
    y: (N, 1) - la cible
    beta: (P, 1) - les coefficients actuels du modèle
    """
    N = X.shape[0]
    
    # 1. Calcul des prédictions du modèle (N, 1)
    y_pred = X @ beta
    
    # 2. Calcul du vecteur d'erreur (N, 1)
    erreur = y_pred - y
    
    # 3. Calcul du gradient (P, 1)
    # X.T est (P, N) et erreur est (N, 1) -> le produit @ donne bien (P, 1)
    gradient = (2 / N) * (X.T @ erreur)
    
    return gradient
```
