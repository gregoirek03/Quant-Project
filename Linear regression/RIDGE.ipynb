```python
import numpy as np

def fit_ridge(X, y, lam=1.0):
    """
    Calcule les coefficients de la régression Ridge.
    """
    n_samples = X.shape[0]
    
    # 1. On ajoute une colonne de 1 à gauche de la matrice X pour l'intercept
    X_biased = np.hstack((np.ones((n_samples, 1)), X))
    n_features = X_biased.shape[1]
    
    # 2. Création de la matrice identité (p x p) pour la pénalité
    I = np.eye(n_features)
    I[0, 0] = 0  # On ne pénalise pas le premier élément (l'intercept)
    
    # 3. Calcul de (X^T * X + lambda * I)
    XtX = np.dot(X_biased.T, X_biased)
    regularized_matrix = XtX + lam * I
    
    # 4. Résolution du système (X^T * X + lambda * I) * beta = X^T * y
    # np.linalg.solve est plus rapide et stable que d'inverser manuellement
    beta = np.linalg.solve(regularized_matrix, np.dot(X_biased.T, y))
    
    return beta

def predict_ridge(X, beta):
    """
    Prédit les valeurs de y à partir de X et des coefficients beta.
    """
    n_samples = X.shape[0]
    X_biased = np.hstack((np.ones((n_samples, 1)), X))
    
    return np.dot(X_biased, beta)
```
