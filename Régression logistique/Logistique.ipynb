```python
import numpy as np

# 1. Configuration et génération de données fictives de classification
np.random.seed(42) #obtenir des valeurs aléatoires
n_samples = 100 #nombre d'échantillons
n_features = 2 #nombre de paramètres

# Données explicatives (X)
X = np.random.randn(n_samples, n_features)

# Génération de X_b (on ajoute la colonne de 1 pour l'intercept beta_0)
X_b = np.c_[np.ones((n_samples, 1)), X]

# Vrais coefficients théoriques (3 lignes pour beta_0, beta_1, beta_2)
true_beta = np.array([[0.5], [2.5], [-1.5]])

# Fonction sigmoïde
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Génération de la cible Y (0 ou 1) en utilisant les vrais coefficients + un bruit
probabilities = sigmoid(np.dot(X_b, true_beta))
Y = (probabilities > 0.5).astype(int)  # astype(int) : Convertit les True/False en 1/0

# 2. Initialisation pour la Descente de Gradient
beta = np.zeros((n_features + 1, 1)) # On part de zéro
learning_rate = 0.1
n_iterations = 1000

# 3. Boucle d'entraînement
for i in range(n_iterations):
    # Prédiction (Probabilités)
    Z = np.dot(X_b, beta)
    Y_pred = sigmoid(Z)
    
    # Calcul du gradient
    gradient = (1 / n_samples) * np.dot(X_b.T, (Y_pred - Y))
    
    # Mise à jour des betas
    beta = beta - learning_rate * gradient
    
    # Optionnel : Afficher l'erreur toutes les 200 itérations
    if i % 200 == 0:
        # Formule de la Log Loss
        loss = - (1 / n_samples) * np.sum(Y * np.log(Y_pred + 1e-15) + (1 - Y) * np.log(1 - Y_pred + 1e-15))
        print(f"Itération {i} -> Erreur (Loss): {loss:.4f}")

print("\nBetas finaux optimisés :\n", beta)
```

```python
# 1. Pourquoi l'erreur démarre précisément à 0.6931 ?
# --> À l'itération 0, l'erreur vaut 0.6931
# --> On a initialisé tous nos coefficients zéro
# --> Notre modèle n'en sait absolument rien et prédit une probabilité parfaite de 0.5 (pile ou face) pour chaque échantillon
# --> sigma(0) = 0.5
# --> On calcule la fonction de coût (Log Loss) pour une prédiction constante de 0.5 
# --> La formule simplifiée donne -ln(0.5) = 0.693147
# --> C'est le comportement idéal : le modèle part d'une incertitude totale (le hasard pur), puis l'erreur chute drastiquement (de 0.69 à 0.12)
# --> Prouve que le modèle apprend efficacement !
```

```python
#2. Pourquoi les estimations des coefficients ne sont pas parfaites ?
# --> Maximiser la certitude de ses prédictions (rapprocher les probabilités au plus près de 1 ou de 0)
# --> Régression logistique cherche à rendre la pente de sa courbe sigmoïde la plus raide possible. 
# --> Pour y arriver, elle fait grimper la taille de ses coefficients beta
```
