```python
# Écrire une fonction normaliser_matrice(X) :
# En entrée : une matrice NumPy X 
# En sortie : la matrice normalisée 

import numpy as np

def normaliser_matrice(X: np.ndarray) -> np.ndarray:
    """
    Normalise une matrice variable par variable (Z-score) de manière vectorisée.
    Gère le cas des colonnes constantes pour éviter la division par zéro.
    """
    # 1. S'assurer que l'entrée est bien un tableau NumPy
    X = np.asarray(X)
    
    # 2. Calculer la moyenne et l'écart-type de chaque colonne (verticalement)
    moyennes = np.mean(X, axis=0)
    ecarts_types = np.std(X, axis=0)
    
    # 3. Gérer la contrainte de la division par zéro :
    # Si un écart-type est égal à 0 (colonne constante), on le remplace temporairement 
    # par 1 pour éviter le crash. On utilise np.where(condition, si_vrai, si_faux)
    ecarts_types_securises = np.where(ecarts_types == 0, 1.0, ecarts_types)
    
    # 4. Application du Z-score grâce au Broadcasting
    # NumPy va soustraire la moyenne de la col A à toutes les lignes de la col A, etc.
    X_normalise = (X - moyennes) / ecarts_types_securises
    
    # 5. Sécurité finale : Si une colonne avait un écart-type de 0, 
    # elle vaut maintenant 0 partout (au lieu de bugger), ce qui est propre.
    X_normalise[:, ecarts_types == 0] = 0.0
    
    return X_normalise

# Remarque : 
# axis=0 : Tu dis à NumPy de faire le calcul à travers les lignes (du haut vers le bas). 
# --> écrase tes 100 lignes pour obtenir 1 seule ligne contenant 3 moyennes (une moyenne pour chaque variable). 

# sinon axis=1
```
