```python
# Calculer le rendement quotidien
# La formule mathématique du rendement au jour $t$ est :
# $$R_t = \frac{\text{Prix}_t - \text{Prix}_{t-1}}{\text{Prix}_{t-1}}$$

# ❌ Niveau 1 : Le réflexe du débutant (La boucle for)

import pandas as pd
import numpy as np

# On crée un faux historique de 5 jours pour l'exemple
df = pd.DataFrame({'Prix': [100, 102, 101, 105, 104]})

rendements = [np.nan] # Le premier jour n'a pas de jour précédent, donc pas de rendement

for i in range(1, len(df)):
    rendement_jour = (df['Prix'][i] - df['Prix'][i-1]) / df['Prix'][i-1]
    rendements.append(rendement_jour)

df['Rendement_Boucle'] = rendements
```

```python
# Niveau 2 : Le premier pas en vectorisation (Le "Shift")
# 
# Pour vectoriser, on doit éliminer la notion de "temps" et de "boucle". 
# On veut aligner le prix d'aujourd'hui et le prix d'hier sur la même ligne pour que le processeur fasse le calcul d'un coup.

# Pour ça, on utilise .shift(1) qui décale toutes les données d'une ligne vers le bas.

# .shift(1) crée une colonne avec le prix de la veille
df['Prix_Veille'] = df['Prix'].shift(1)

# Calcul vectorisé direct sur toute la colonne
df['Rendement_Shift'] = (df['Prix'] - df['Prix_Veille']) / df['Prix_Veille']
```

```python
# 🏆 Niveau 3 : Le niveau Quant Pro (pct_change)
# .pct_change() (Percentage Change)
# --> utilisable directement

# L'arme absolue du quant : propre, rapide, vectorisé en C
df['Rendement_Master'] = df['Prix'].pct_change()
```

```python
# Application : 
# créer un signal de trading simple basé sur ces rendements :
# Si le rendement du jour est positif, on achète (on note 1)
# S'il est négatif ou nul, on reste hors du marché (on note 0).

df['Signal'] = np.where(df['Rendement_Master'] > 0, 1, 0)

```
